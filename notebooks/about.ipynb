```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure
fig, ax = plt.subplots(figsize=(14, 10))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.axis("off")

# Define positions for components
positions = {
    "Receiver": (5, 9),
    "Self-Agent": (5, 7.5),
    "Conscious Agent": (7.5, 6),
    "Unconscious Agent": (2.5, 6),
    "Relationship Manager": (8.5, 4.5),
    "Memory Cache": (5, 4.5),
    "Retriever": (2.5, 3),
    "Flywheel": (5, 1.5),
}

# Draw nodes
for name, (x, y) in positions.items():
    ax.add_patch(mpatches.FancyBboxPatch(
        (x-1, y-0.5), 2, 1, boxstyle="round,pad=0.1",
        edgecolor="black", facecolor="lightblue", lw=2))
    ax.text(x, y, name, ha="center", va="center", fontsize=11, weight="bold")

# Helper function for arrows
def connect(start, end, text=None, style="simple", color="black"):
    ax.annotate("", xy=positions[end], xytext=positions[start],
                arrowprops=dict(arrowstyle="->", lw=1.5, color=color))
    if text:
        sx, sy = positions[start]
        ex, ey = positions[end]
        ax.text((sx+ex)/2, (sy+ey)/2+0.2, text, ha="center", va="center", fontsize=9, color=color)

# Connections
connect("Receiver", "Self-Agent", "normalized input")
connect("Self-Agent", "Conscious Agent", "probe / reasoning")
connect("Conscious Agent", "Flywheel", "conceptual updates", color="blue")
connect("Self-Agent", "Flywheel", "representation update", color="blue")
connect("Unconscious Agent", "Flywheel", "background updates", color="blue")
connect("Flywheel", "Retriever", "meta-learning sync", color="purple")
connect("Retriever", "Conscious Agent", "relevant knowledge", color="green")
connect("Retriever", "Unconscious Agent", "patterns / heuristics", color="green")
connect("Memory Cache", "Conscious Agent", "working memory")
connect("Memory Cache", "Unconscious Agent", "fast memory")
connect("Self-Agent", "Memory Cache", "short-term store", color="brown")
connect("Receiver", "Relationship Manager", "interaction context")
connect("Relationship Manager", "Conscious Agent", "tone / state")
connect("Relationship Manager", "Self-Agent", "context awareness")

# Title
ax.set_title("Layered Cognitive Architecture: Data Flow & Feedback Loops", fontsize=14, weight="bold")

plt.show()
```
